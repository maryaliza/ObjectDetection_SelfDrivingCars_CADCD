# **Object Detection**

```python
import cv2 as cv
font=cv2.FONT_HERSHEY_COMPLEX


net = cv.dnn_DetectionModel(frozen_model_path, config_file_path)
net.setInputSize(320, 320)
net.setInputScale(1.0 / 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)

img = cv2.imread(path)
frame = img

classes, confidences, boxes = net.detect(frame, confThreshold=0.4)

for classId, confidence, box in zip(classes.flatten(), confidences.flatten(), boxes):
    confdnc=float("{:.2f}".format(confidence))
    print(Class_lables[classId-1], confdnc)
    cv.rectangle(frame, box, color=(0,0,255))
    cv.putText(frame,Class_lables[classId-1],(box[0],box[1]),font,1,(0,0,0),1)

# cv.imshow('out', frame)
# cv.waitKey()
```
